```python
def find_key(item1, item2):
    splitted1 = item1.split('-')
    splitted2 = item2.split('-')
    length = min(len(splitted1), len(splitted2))
    ind = 0
    while ind < length:
        if splitted1[ind] != splitted2[ind]:
            break;
        ind+=1
    key = "-".join(splitted1[:ind])
    return key
```

```python
find_key("o-ssad-2131-zxc-asaxzc--xzc-cxczxc-zc-cxzc--ad","o-ssad-2131-zasddad-asd-asd--dadd")
```

```python
def get_data(key, name):
    song_name = name.replace(key, "")
    song_name = " ".join(song_name.split('-')[:-1])
    artist = " ".join(key.split('-'))
    return artist, song_name
```

```python
get_data('J-Cole','J-Cole-Shes-Mine-Pt-2-458506.json')
```

```python
def normalize_paths(paths):
    res = []
    key = paths[0]
    
    length = len(paths)
    i = 1
    while i < length:
        x = paths[i]
        x = x.replace('--', '-')
        if not x[0].isalpha() or '%' in x:
            i += 1
            continue
        if key not in x:
            key = find_key(x, paths[i+1])
        if len(key) <= 2:
            i+=1
            key = 'fail'
            continue;
            
        new_data = get_data(key, x)
        res.append((new_data[0], new_data[1], paths[i]))
        i+=1
    return res
```

```python
res = normalize_paths(lyrics_files)
```

```python
def fix_lyrics(lyrics):
    res = []
    for x in lyrics:
        if "<li style=" in x:
            continue
        x = x.replace("<li>", "")
        x = x.replace("</li>", "")
```

```python
total = 60000
cnt = 1
batch_cnt = 1
batch_size = 10000

batch_dicts = []
for el in res:
    with open('lyrics/texts/' + el[2], 'r') as file:
        try:
            lyrics = json.load(file)
        except:
            continue
        if "<li>We do not have the lyric for this song yet.</li>" in lyrics:
            continue
        fixed_lyrics = fix_lyrics(lyrics)
        clear_output()
        print(cnt)
        batch_dicts.append({
            'id': str(batch_cnt) + '_' + str(cnt),
            'author': el[0],
            'song': el[1],
            'lyrics': fixed_lyrics
        })
        cnt += 1
    if cnt == batch_size:
        with open(batch_dicts[-1]['id'] + "_batch.json", 'w') as file:
            json.dump(batch_dicts, file)
        batch_cnt+= 1
        cnt = 1
```
